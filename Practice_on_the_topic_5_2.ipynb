```python
import pandas as pd
```

```python
df = pd.read_csv(r'gdaSet.csv')
df.head(10)
```

```python
df = df.drop_duplicates()
df.head(10)
```

```python
df = df.drop(columns =['Email'])
df
```

```python
#modifying a column value
df['Job Title'] = df['Job Title'].astype(str).replace({'Games developer': 'IT Assistant'})
df
```

```python
def generate_address(index):
  return f"1234 Nairobi st patricks{index}, AN 1234"
df['Adress'] = df.index.map(generate_address)
df
```

```python
df = df.dropna()
df.head()
#fillna
```

```python
#data sorting
df.describe()
```

```python
sorted_value = df.sort_values(by='Age', ascending=False)
sorted_value
```

```python
#GROUPING
grouped_df = df.groupby('Sex')
grouped_df['Salary'].sum()
```

```python
#AGGREGATE FUNCTIONS
grouped_df[['Age', 'Salary']].sum()
```
