```python
#loading and reading data
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

data=pd.read_csv('/content/supermarket_sales - Sheet1.csv')
print(data.head(10))
print("\n" + "-"*50 + "\n")

print(data.info)
print("\n" + "-"*50 + "\n")
#view collums

print(data.columns)
print("\n" + "-"*50 + "\n")
```

```python
#data cleaning and formating
#remove missing values
sales_data=data.dropna()
print(sales_data.head)
```

```python
# mathematical statistics for data processing
data_gender=sales_data.groupby('Gender')['Total'].mean()
print(data_gender)
print("\n" + "-"*50 + "\n")
data_product=sales_data.groupby('Product line')['Total'].mean()
print(data_product)
print("\n" + "-"*50 + "\n")
data_payment=sales_data.groupby('Payment')['Total'].mean()
print(data_payment)
print("\n" + "-"*50 + "\n")
data_customer=sales_data.groupby('Customer type')['Total'].mean()
print(data_customer)
print("\n" + "-"*50 + "\n")
sales_total=sales_data['Total'].sum()
print(sales_total)
```

```python
from typing import Pattern
#search for patterns in the data was performed.
Pattern=sales_data.groupby('Product line')['Total'].mean()
print(Pattern)
```

```python
#data visualisation

sns.boxplot(x='Gender',y='Product line', data=sales_data)
plt.show()
print("\n" + "-"*50 + "\n")

sns.lineplot(data=sales_data, hue='Gender', x='Product line', y='Total')
plt.show()
print("\n" + "-"*50 + "\n")

plt.scatter(sales_data['Product line'],sales_data['Total'])
plt.show()
print("\n" + "-"*50 + "\n")

```

```python
from math import hypot

#hypothesis testing
import scipy.stats as stats
ci = stats.t.interval(0.95, len(sales_data)-1, loc=np.mean(sales_data['Total']), scale=stats.sem(sales_data['Total']))
hypothesis_test=sales_data.groupby('Gender')['Total'].mean()
print(hypothesis_test)
print(ci)
```
